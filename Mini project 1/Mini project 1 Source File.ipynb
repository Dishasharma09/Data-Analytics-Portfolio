#  Data Cleaning

To ensure accurate analysis, the dataset is cleaned by:

- Removing duplicate records
- Converting numerical columns
- Converting date columns
- Handling invalid values

```python
df = df.drop_duplicates()
```

```python
df["Patient Waittime"] = pd.to_numeric(
    df["Patient Waittime"],
    errors="coerce"
)

df["Patient Satisfaction Score"] = pd.to_numeric(
    df["Patient Satisfaction Score"],
    errors="coerce"
)
```

```python
df["Patient Admission Date"] = pd.to_datetime(
    df["Patient Admission Date"],
    format="%d-%m-%Y %H:%M",
    errors="coerce"
)
```

```python
df.isnull().sum()
```

```python
df.head()
```

# Feature Engineering

Feature Engineering is the process of creating new variables from existing data to improve analysis.

In this project, new features such as Month and Age Group are created to better understand patient admission trends and demographic distribution.

```python
# Create Month column

df["Month"] = df["Patient Admission Date"].dt.month_name()
```

```python
# Create Age Group column

bins = [0,18,35,50,65,100]

labels = [
    "0-18",
    "19-35",
    "36-50",
    "51-65",
    "65+"
]

df["Age Group"] = pd.cut(
    df["Patient Age"],
    bins=bins,
    labels=labels
)
```

```python
df.head()
```

Verify that the newly created features have been added successfully before proceeding with the analysis.

# Key Performance Indicators (KPIs)

Key Performance Indicators provide a quick overview of the Emergency Room's overall performance.

The following metrics summarize patient volume, waiting time, satisfaction, department workload and admission status.

```python
total_patients = len(df)

avg_wait = df["Patient Waittime"].mean()

avg_satisfaction = df["Patient Satisfaction Score"].mean()

total_departments = df["Department Referral"].nunique()

total_admissions = (
    df["Patient Admission Flag"]=="Admitted"
).sum()

admission_rate = (
    total_admissions/total_patients
)*100
```

```python
print("Total Patients :", total_patients)

print("Average Waiting Time :", round(avg_wait,2),"Minutes")

print("Average Satisfaction Score :", round(avg_satisfaction,2))

print("Total Departments :", total_departments)

print("Total Admissions :", total_admissions)

print("Admission Rate :", round(admission_rate,2),"%")
```

# Bivariate and Multivariate Analysis

While Univariate Analysis examines a single variable at a time, Bivariate and Multivariate Analysis explore relationships between two or more variables.

These analyses help identify patterns that cannot be observed by studying variables individually.

The objective is to understand how patient demographics, waiting time, department referrals and satisfaction scores influence each other.

## Business Question 18

### Which departments have the highest average patient waiting time?

```python
avg_wait = (
    df.groupby("Department Referral")["Patient Waittime"]
      .mean()
      .sort_values(ascending=False)
)

avg_wait
```

```python
plt.figure(figsize=(12,6))

avg_wait.plot(kind="bar")

plt.title("Average Waiting Time by Department")

plt.xlabel("Department")

plt.ylabel("Average Waiting Time (Minutes)")

plt.xticks(rotation=45)

plt.show()
```

### Key Findings

- Average waiting time varies across hospital departments.
- Some departments consistently experience longer patient queues.
- Waiting time differences may indicate workload imbalance or staffing limitations.

### Business Recommendation

- Increase staffing in departments with the highest average waiting times.
- Monitor patient flow continuously to reduce delays.

## Business Question 19

### Which departments achieve the highest average patient satisfaction score?

```python
avg_sat = (
    df.groupby("Department Referral")["Patient Satisfaction Score"]
      .mean()
      .sort_values(ascending=False)
)

avg_sat
```

```python
plt.figure(figsize=(12,6))

avg_sat.plot(kind="bar")

plt.title("Average Satisfaction Score by Department")

plt.xlabel("Department")

plt.ylabel("Average Satisfaction Score")

plt.xticks(rotation=45)

plt.show()
```

### Key Findings

- Patient satisfaction differs across departments.
- High-performing departments can serve as benchmarks for operational improvements.

### Business Recommendation

- Identify best-performing departments and replicate their service practices across the hospital.

## Business Question 24

### Is there a correlation among patient age, waiting time and satisfaction score?

```python
corr = df[
    [
        "Patient Age",
        "Patient Waittime",
        "Patient Satisfaction Score"
    ]
].corr()

corr
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

### Key Findings

- The heatmap displays the strength and direction of relationships among numerical variables.
- Correlation values close to zero indicate weak relationships.

### Business Recommendation

- Since patient satisfaction may depend on multiple factors, hospitals should analyze additional variables beyond waiting time.
