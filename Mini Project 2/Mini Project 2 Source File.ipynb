# Outlier Detection Strategy

Financial loss values may contain extreme values that can distort analysis.

We use boxplots and IQR method to detect outliers.

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Amount_Lost_INR'])
plt.title("Outlier Detection - Amount Lost")
plt.show()
```

# IQR Method Explanation

We calculate:

- Q1 (25th percentile)
- Q3 (75th percentile)
- IQR = Q3 - Q1

Then define:

- Lower Bound = Q1 - 1.5 * IQR
- Upper Bound = Q3 + 1.5 * IQR

```python
Q1 = df['Amount_Lost_INR'].quantile(0.25)
Q3 = df['Amount_Lost_INR'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

# Outlier Handling Strategy

Outliers will be treated using capping method:

- Values below lower bound → replaced with lower bound
- Values above upper bound → replaced with upper bound

```python
df['Amount_Lost_INR'] = np.where(
    df['Amount_Lost_INR'] < lower_bound,
    lower_bound,
    np.where(
        df['Amount_Lost_INR'] > upper_bound,
        upper_bound,
        df['Amount_Lost_INR']
    )
)
```

# Cyber Risk Segmentation Strategy

We classify incidents into risk levels based on financial impact to support prioritization.

```python
def risk_level(x):
    if x < 100000:
        return "Low"
    elif x < 250000:
        return "Medium"
    elif x < 400000:
        return "High"
    else:
        return "Critical"

df['Risk_Level'] = df['Amount_Lost_INR'].apply(risk_level)

df['Risk_Level'].value_counts()
```

# Risk Level Distribution

This shows how incidents are distributed across Low, Medium, High, and Critical risk categories.

```python
sns.countplot(data=df, x='Risk_Level', order=['Low','Medium','High','Critical'])
plt.title("Cyber Risk Distribution")
plt.show()
```
